```python
data=dict(type='choropleth',
          colorscale='Portland',
          locations=us['code'],
          z=us['total exports'],
          locationmode='USA-states',
          text=us['text'],
          colorbar={'title':'Millions USD'}) 
#text variable hovering over the field to display information
```

```python
layout=dict(title='2011 US Agriculture Exports by States',
           geo=dict(scope='usa',
                   showlakes=True
            ))
```

```python
#should be connected o the net for this to work
chloromap=dict(data=[data],layout=layout)
iplot(chloromap)
```

```python
#World Map
```

```python
data=dict(type='choropleth',
          colorscale='Portland',
          locations=world['CODE'],
          z=world['GDP (BILLIONS)'],
          text=world['COUNTRY'],
          colorbar={'title':'Millions USD'}) 
```
