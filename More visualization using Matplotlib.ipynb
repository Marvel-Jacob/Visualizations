```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.plotly as py
```

```python
tel=pd.read_csv('C:/Users/Administrator/Desktop/Data visualization/datasets/Telco-Customer-Churn.csv')
```

```python
#Single pie

pmt_method=tel.groupby(['PaymentMethod']).size()
pmt_method=pd.DataFrame(pmt_method)
pmt_method=pmt_method.reset_index()
pmt_method=pmt_method.rename(columns={0:'Count'})
```

```python
col=['purple','indigo','cyan','plum']
plt.pie(pmt_method['Count'],labels=pmt_method['PaymentMethod'],labeldistance=0.9,
       autopct='%0.1f%%',startangle=90,wedgeprops=dict(width=0.4),radius=1)

plt.title("Variation by Payment Method",loc='center',color='khaki')
plt.axis('equal')
plt.show()
```

```python
#double pie
pmt_gender=tel.groupby(['PaymentMethod','gender']).size()
pmt_gender=pd.DataFrame(pmt_gender)
pmt_gender=pmt_gender.reset_index()
pmt_gender=pmt_gender.rename(columns={0:'Count'})
```

```python
col=['coral','crimson','peru','hotpink']
plt.pie(pmt_gender[pmt_gender['gender']=='Male']['Count'],
       autopct='%0.1f%%',pctdistance=0.8,startangle=90,
       wedgeprops=dict(width=0.5),radius=1,colors=col)
plt.pie(pmt_gender[pmt_gender['gender']=='Female']['Count'],
       autopct='%0.1f%%',pctdistance=0.8,startangle=90,
       wedgeprops=dict(width=0.5),radius=1.55,colors=col)
plt.title("Variation by Payment Method",loc='center')
plt.legend(labels=pmt_gender[pmt_gender['gender']=='Female']['PaymentMethod'],
          loc='right',bbox_to_anchor=(1.4,0.7))
plt.axis('equal')
plt.show()
```
