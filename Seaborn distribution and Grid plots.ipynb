```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
tips=sns.load_dataset('tips')
tips.head(5)
```

```python
#run it twice
sns.distplot(tips['total_bill'])
mpl.rcParams['patch.force_edgecolor']=True
mpl.rcParams['patch.edgecolor']= "Black"
```

```python
plt.figure(figsize=(10,7))
sns.distplot(tips['tip'],kde=False,color='navy')
mpl.rcParams['patch.force_edgecolor']= True
mpl.rcParams['patch.edgecolor']= "Black"
plt.title('Distribution of Tips')
plt.ylabel('Frequency')
plt.show()
```

```python
#joint plot gives the scatter and distribution


sns.jointplot(x="total_bill",y='tip',data=tips, kind='scatter', size=5)
plt.show()

sns.jointplot(x="total_bill",y='tip',data=tips, kind='kde', size=5)
plt.show()

sns.jointplot(x="total_bill",y='tip',data=tips, kind='hex',size=5)
plt.show()

sns.jointplot(x="total_bill",y='tip',data=tips, kind='reg', size=5)
plt.show()
```

```python
#pear plot picks the numerical variables and finds the relation between them

#hue can be given as a catgorical variable
#(multivariant)

sns.pairplot(tips,hue='smoker')
plt.show()
```

```python
sns.kdeplot(tips['tip'],legend=True,shade=True)
plt.show()
```

```python
#using iris dataset
```

```python
iris=sns.load_dataset('iris')
iris.head(10)
```

```python
#creates a grid like structure for various numerical plots

grid=sns.PairGrid(iris,hue='species',palette='inferno')
grid.map(plt.scatter).add_legend()
plt.show()
```

```python
grid=sns.PairGrid(iris,hue='species',palette='inferno')
grid.map_diag(plt.hist)
grid.map_lower(plt.scatter)
grid.map_upper(sns.kdeplot)
grid.add_legend()
plt.show()
```

```python
#working on tips datset

#facet grid is for inputting rows and columns(for 2 categorical variable)

grid=sns.FacetGrid(tips,row='smoker',col='time')
grid.map(plt.hist,'total_bill')
plt.show()
```

```python
grid=sns.FacetGrid(tips,row='smoker',col='time',hue='sex')
grid.map(plt.scatter,'total_bill','tip')
plt.show()
```

```python
grid=sns.JointGrid(x='total_bill',y='tip',data=tips,dropna=True)
grid.plot(plt.scatter,sns.distplot)
plt.show()
```

```python
f,axes=plt.subplots(1,2,figsize=(6,4))
sns.boxplot(y='total_bill',x='sex',data=tips,ax=axes[0])
sns.boxplot(y='total_bill',x='smoker',data=tips,ax=axes[1])
sns.set_style('darkgrid')
plt.show()
```

```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm')
plt.show()
```
