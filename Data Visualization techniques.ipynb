```python
import matplotlib.pyplot as plt #pyploy used for 2D graphing
import numpy as np
```

```python
x=np.linspace(0,9,10) #by default float value
x
```

```python
y=x**2
y
```

```python
#if u dont specify anyting then itll b a line chart

plt.plot(x,y)
plt.show()
```

```python
#for putting labels

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line graph')
plt.plot(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.subplot(2,2,1) #(rows,columns,subplot index)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line graph')
plt.plot(x,y,'r')

plt.subplot(2,2,2)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line graph')
plt.plot(y,x,'g')

plt.subplot(2,2,3)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line graph')
plt.plot(x**2,y,'b')

plt.subplot(2,2,4)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line graph')
plt.plot(-y,-x,'y')

plt.tight_layout()

plt.show()
```

```python
#the object oriented way

fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y,'g--') #colorpattern
plt.show()
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.9,0.9])
axes2=fig.add_axes([0.2,0.5,0.4,0.4]) #([x length,y length, x size, y size])
axes1.plot(x,y,'g--') #colorpattern
axes2.plot(y,x,'r*-')
plt.show()
```

```python
fig,axes=plt.subplots(1,2)
axes[0].plot(x,y,'r*-')
axes[1].plot(y,x,'g*-')

axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].set_title('X-Y relation')

axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
axes[1].set_title('Y-X relation')

plt.show()
```

```python
fig,axes=plt.subplots(2,2)
axes[0,0].plot(x,y)
axes[0,1].plot(y,x)
axes[1,0].plot(x,y)
axes[1,1].plot(y,x)
plt.show()
```

```python
fig=plt.figure()
plt.plot(x,y,'r--',label='x-y')
plt.plot(y,x,'g--',label='y-x')
plt.legend(loc=0)

plt.savefig("xy and yx graphs.jpeg") #to save the file to ur directory

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
tenure=pd.read_csv('C:/Users/Administrator/Desktop/Data visualization/datasets/Telco-Customer-Churn.csv')
tenure
```

```python
tenure.columns
```

```python
#1. how is the distribution of tenure

plt.hist(tenure['tenure'],edgecolor='black',color='red')
plt.title('Distribution for tenure')
plt.xlabel('tenure')
plt.ylabel('values')
plt.show()
```

```python
#2. how is the distribution of monthly charges

plt.hist(tenure['MonthlyCharges'],edgecolor='black',color='chartreuse')
plt.title('Distribution for Monthly charges')
plt.xlabel('monthly charges')
plt.ylabel('values')
plt.show()
```

```python
#3. What is the part of the whole relationship of internet services

p=tenure.groupby('InternetService').size()
p=pd.DataFrame(p)
p=p.reset_index()
p=p.rename(columns={0:'count'})
con

plt.pie(p['count'],labels=p['InternetService'],autopct='%1.1f%%',startangle=90)
plt.axis('equal')
plt.show()
```

```python
#4. contract which is most common and the perc of the most common contract

con=tenure.groupby('Contract').size()
con=pd.DataFrame(con)
con=con.reset_index()
con=con.rename(columns={0:'count'})
con

plt.pie(con['count'],labels=con['Contract'],autopct='%1.1f%%',startangle=90)
plt.axis('equal')
plt.show()
```

```python
#5. What is the part to whole relationship of multiplelines

ML=tenure.groupby('MultipleLines').size()
ML=pd.DataFrame(ML)
ML=ML.reset_index()
ML=ML.rename(columns={0:'count'})
ML

plt.pie(ML['count'],labels=ML['MultipleLines'],autopct='%1.1f%%',startangle=90,explode=(0.05,0.05,0.05))
plt.axis('equal')
plt.show()
```

```python
#6. what is the ratio of churn to not churn

cyes=tenure.groupby('Churn').size()
cyes=pd.DataFrame(cyes)
cyes=cyes.reset_index()
cyes=cyes.rename(columns={0:'count'})
cyes

plt.pie(cyes['count'],labels=cyes['Churn'],autopct='%1.1f%%',startangle=90,explode=(0.1,0.1))
plt.axis('equal')
plt.show()

ratio=73.5/26.5
print("The ratio is",ratio)
```

```python
#7. What is the ratio of senior citizens

sc=tenure.groupby('SeniorCitizen').size()
sc=pd.DataFrame(sc)
sc=sc.reset_index()
sc=sc.rename(columns={0:'count'})
sc

plt.pie(sc['count'],labels=sc['SeniorCitizen'],autopct='%1.1f%%',startangle=90,explode=(0.3,0.1))
plt.axis('equal')
plt.show()

ratio=83.8/16.2
print("The ratio is",ratio)
```

```python
#1. which season is  most popular

plt.hist(ath['Season'],color='turquoise',edgecolor='black')
plt.show()
```

```python
#2. which are the cities which hosted the olympics? which city has hosted the max number of olympics

unique_cities=ath['City'].unique()
unique_cities=pd.DataFrame(unique_cities)
unique_cities
#print(unique_cities)

dat=ath[['City','Year']]
dat.groupby(['Year','City']).count()


#plt.hist(ath['City'],egdecolor='black',color='teal')
#plt.show()
```

```python
#3. top 10 teams having numbers of athelete

no_ath=ath[['Name','Team']]
no_ath=no_ath.groupby('Team').count()
no_ath=no_ath.reset_index()
no_ath=no_ath.rename(columns={'Name':'Count of atheletes'})
no_ath=no_ath.sort_values(by='Count of atheletes',ascending=False)
no_ath=no_ath.head(10)
no_ath

plt.bar(no_ath['Team'],no_ath['Count of atheletes'],color='crimson',edgecolor='black')
plt.xticks(rotation=90)
plt.title('top 10 teams having no of players')
plt.ylabel('no: of players')
plt.xlabel('Teams')
plt.show()

    
```

```python
#4. top 10 teams having max avg height of atheletes

h_ath=ath[['Team','Height']]
h_ath=h_ath.dropna(subset=['Height'])
h_ath=h_ath.groupby('Team').mean()
h_ath=h_ath.reset_index()
h_ath=h_ath.sort_values(by='Height',ascending=False)
h_ath=h_ath.head(10)
print(h_ath)

plt.bar(h_ath['Team'],h_ath['Height'],color='coral',edgecolor='black')
plt.xticks(rotation=90)
plt.title('top 10 teams having max height of players')
plt.ylabel('height')
plt.xlabel('Teams')
plt.show()


```

```python
#5. top 10 teams having max avg weight of athelets

w_ath=ath[['Team','Weight']]
w_ath=w_ath.dropna(subset=['Weight'])
w_ath=w_ath.groupby('Team').mean()
w_ath=w_ath.reset_index()
w_ath=w_ath.sort_values(by='Weight',ascending=False)
w_ath=w_ath.head(10)
print(w_ath)

plt.bar(w_ath['Team'],w_ath['Weight'],color='plum',edgecolor='black')
plt.xticks(rotation=90)
plt.title('top 10 teams having max weight of players')
plt.ylabel('weight')
plt.xlabel('Teams')
plt.show()
```

```python
#6. top 10 sports which have max participation
max_part=ath[['ID','Sport']]
max_part=max_part.groupby('Sport').count()
max_part=max_part.reset_index()
max_part=max_part.rename(columns={'ID':'count'})
max_part=max_part.sort_values(by='count',ascending=False)
max_part=max_part.head(10)
#print(h_ath)
max_part


plt.bar(max_part['Sport'],max_part['count'],color='navy',edgecolor='black')
plt.xticks(rotation=90)
plt.title('top 10 sport having max participation')
plt.ylabel('times played')
plt.xlabel('Sports')
plt.show()
```

```python
#7. top 10 teams having recieved max gold
gold=ath.query('Medal=="Gold"')
top_teams=gold[['ID','Team']]
top_teams=top_teams.groupby('Team').count()
top_teams=top_teams.reset_index()
top_teams=top_teams.rename(columns={'ID':'count'})
top_teams=top_teams.sort_values(by='count',ascending=False)
top_teams=top_teams.head(10)
top_teams

plt.bar(top_teams['Team'],top_teams['count'],color='wheat',edgecolor='black')
plt.xticks(rotation=90)
plt.title('top 10 teams having max gold')
plt.ylabel('received golds')
plt.xlabel('Team')
plt.show()
```

```python
#8. top 10 countries having received max silver. Do u see any difference in the countries with gold and silver?

gold=ath.query('Medal=="Silver"')
top_teams1=gold[['ID','Team']]
top_teams1=top_teams1.groupby('Team').count()
top_teams1=top_teams1.reset_index()
top_teams1=top_teams1.rename(columns={'ID':'count'})
top_teams1=top_teams1.sort_values(by='count',ascending=False)
top_teams1=top_teams1.head(10)
top_teams1

plt.bar(top_teams1['Team'],top_teams1['count'],color='wheat',edgecolor='black')
plt.xticks(rotation=90)
plt.title('top 10 teams having max silver')
plt.ylabel('received silvers')
plt.xlabel('Team')
plt.show()
```
